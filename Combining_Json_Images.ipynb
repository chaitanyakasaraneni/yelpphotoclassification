```python
# Importing libraries we will need for our model building.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from glob import glob
import seaborn as sns
from PIL import Image
np.random.seed(123)
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import itertools
```

```python
#reading json files
photos_json = 'photo.json'
df = pd.read_json(photos_json, lines=True)
```

```python
# creating dictionary with key as photo_id and value as image path
import glob
img_dir = '/Users/supriyajain/Desktop/257/Project/yelp_photos/photos/smalldataset/' # Enter Directory of all images 

data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
imageid_path_dict = {os.path.splitext(os.path.basename(x))[0]: x 
                     for x in glob.glob(data_path)}
```

```python
imageid_path_dict
```

```python
#converting labels to categorical variables
df['class_names_label'] = pd.Categorical(df['label']).codes
```

```python
df.head(3)
```

```python
# for each photo_id getting the corresponding path from the dictionary
df['path'] = df['photo_id'].map(imageid_path_dict.get)
```

```python
print(df.dtypes)
```

```python
df = df.dropna()
```

```python
# getting an image from the path
df['image'] = df['path'].map(lambda x: np.asarray(Image.open(x)))
```
